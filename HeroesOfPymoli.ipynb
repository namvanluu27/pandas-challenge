```python
# Top Spenders #
 
# Run basic calculations to obtain the results in the table below
# SN
spender_groupby = purchase_data.groupby("SN")


# Purchase Count
spender_purchase_count = spender_groupby["Purchase ID"].count()


# Average Purchase Price
spender_purchase_price = spender_groupby["Price"].mean()


# Total Purchase Value
spender_purchase_value = spender_groupby["Price"].sum()


# Create a summary data frame to hold the results
top_spenders = pd.DataFrame({"Purchase Count": spender_purchase_count,
                             "Average Purchase Price": spender_purchase_price,
                             "Total Purchase Value": spender_purchase_value})

spenders_clean_table_df = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()

spenders_clean_table_df.style.format({"Average Purchase Total": "${:,.2f}",
                                       "Average Purchase Price": "${:,.2f}",
                                       "Total Purchase Value": "${:,.2f}"})
```

```python
#  Most Popular Items #

 
# Retrieve the Item ID, Item Name, and Item Price columns

# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


# Item ID
items = purchase_data[["Item ID", "Item Name", "Price"]]


# Item group
items_groupby = items.groupby(["Item ID", "Item Name"])


# Purchase Count
items_purchase_count = items_groupby["Price"].count()
             
                                    
# Total Purchase Value
items_purchase_value = items_groupby["Price"].sum()                                  
                                    
                                    
# Item Price
items_price = items_purchase_value / items_purchase_count



# Create a summary data frame to hold the results
most_popular_items_df = pd.DataFrame({"Purchase Count": items_purchase_count,
                                      "Item Price": items_price,
                                      "Total Purchase Value": items_purchase_value})

items_clean_df = most_popular_items_df.sort_values(["Purchase Count"], ascending=False).head()

items_clean_df.style.format({"Item Price": "${:,.2f}",
                             "Total Purchase Value": "${:,.2f}"})
```

```python
# Most Profitable Items
 
# Sort the above table by total purchase value in descending order
items2_clean_df = most_popular_items_df.sort_values(["Total Purchase Value"],
                                                   ascending=False).head()
# Format with currency style
items2_clean_df.style.format({"Item Price":"${:,.2f}",
                              "Total Purchase Value":"${:,.2f}"})
```
